```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy.random as nr

%matplotlib inline
```

```python
df_train = pd.read_csv('train_values.csv')
df_labels = pd.read_csv('train_labels.csv')

df_test = pd.read_csv('test_values.csv')

# add label to train dataframe
df_train["accepted"] = df_labels["accepted"]

# df_train = pd.read_csv('preped/preped_data2_feature_engineered.csv')
# df_test = pd.read_csv('preped/preped_test_&_featured.csv')

print(df_train.shape)
df_train.head()
```

```python
print(df_test.shape)
df_test.head()
```

```python
# Convert all columns with missing values represented as -1 to NaN
cols = ['state_code', 'county_code', 'msa_md']
for col in cols:
    #train data
    df_train.loc[df_train[col] == -1, col] = np.nan
    #test data
    df_test.loc[df_test[col] == -1, col] = np.nan
```

```python
print("TRAIN DATA\n")
print(df_train.isnull().sum())
print("\n\nTEST DATA\n")
print(df_test.isnull().sum())
```

```python
# Fill missing values with column median
for col in df_train.columns:
    # train data
    df_train.loc[df_train[col].isnull(), col] = df_train[col].median()

# Do the same for Test data
for col in df_test.columns:
    # test data
    df_test.loc[df_test[col].isnull(), col] = df_test[col].median()
```

```python
print("TRAIN DATA\n")
print(df_train.isnull().sum())
print("\n\nTEST DATA\n")
print(df_test.isnull().sum())
```

```python
df_train.to_csv('preped/preped_train_monday_night.csv', index=None)
df_test.to_csv('preped/preped_test_monday_night.csv', index=None)
```
