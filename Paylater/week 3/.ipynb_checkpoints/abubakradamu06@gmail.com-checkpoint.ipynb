```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
clients_df = pd.read_csv('week3_challenge.csv')
clients_df.head(100).T
```

```python
print(clients_df.shape)
clients_df.dtypes
```

```python
clients_df.shape[0] * 0.3
```

```python
clients_df.tail(100).T
```

```python
def show_column_null_counts():
    for col in clients_df.columns:
        print(col, ":     ", clients_df[col].isnull().sum())
        
show_column_null_counts()
```

```python
for col in clients_df.columns:
    if clients_df[col].isnull().sum() > clients_df.shape[0] * 0.3:
        clients_df.drop(col, axis=1, inplace = True)
    
```

```python
clients_df.shape
```

```python
int_df = clients_df.select_dtypes(include = ['int64'])
float_df = clients_df.select_dtypes(include = ['float'])

client_num_df = float_df

for col in int_df.columns:
    client_num_df[col] = int_df[col]
    
client_num_df.shape
```

```python
client_num_df.dtypes
```

```python
client_num_df.fillna(0, inplace=True)
```

```python
x = StandardScaler().fit_transform(client_num_df)
```

```python
x
```

```python
pca = PCA(n_components=5)
```

```python
principalComponents = pca.fit_transform(x)
```

```python
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2', 'principal component 3', 'principal component 4', 'principal component 5'])
```

```python
principalDf.head(10)
```

```python
pca.explained_variance_ratio_.sum()
```

```python
model = KMeans(n_clusters=3)
```

```python
model.fit(principalDf)
```

```python
clusters = model.predict(principalDf)
```

```python
clusters
```

```python
client_num_df['clusters'] = clusters
```

```python
client_num_df.groupby(['clusters']).mean()
```
