```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import numpy.random as nr
import math

%matplotlib inline
```

```python
AWS_df = pd.read_csv('AWS_Dataset_Preped.csv')
AWS_df.head(20)
```

```python
print(AWS_df.shape)
print(AWS_df.CustomerID.unique().shape)
```

```python
AWS_df.drop_duplicates(subset = 'CustomerID', keep = 'first', inplace = True)
print(AWS_df.shape)
print(AWS_df.CustomerID.unique().shape)
```

```python
print(AWS_df.loc[AWS_df['Title'] == 'NaN', 'Title'])
print(AWS_df.loc[AWS_df['Suffix'] == 'NaN', 'Suffix'])
```

```python
# Drop columns with too many missing values
AWS_df.drop('Title', axis = 1, inplace = True)
AWS_df.drop('Suffix', axis = 1, inplace = True)
AWS_df.drop('MiddleName', axis = 1, inplace = True)
AWS_df.drop('AddressLine2', axis = 1, inplace = True)
AWS_df.drop('dummy', axis = 1, inplace = True)
```
